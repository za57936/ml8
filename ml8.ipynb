```python
import nltk   #library for NLP
import matplotlib.pyplot as plt  #importing matplotlib library
import urllib.request    #library extracting text from webpage
from bs4 import BeautifulSoup     #library for parsing html & xml texts
```

```python
response = urllib.request.urlopen('http://php.net/')     #request to get web page content
html = response.read()       #response from request
print (html)        #printing content
```

```python
soup = BeautifulSoup(html,"html5lib")   #cleaning all tags of html content & storing in soup variable using BeautifulSoup
text = soup.get_text(strip=True)      #getting all cleaned text into text variable
print (text)      #printing
```

```python
tokens = [t for t in text.split()]   #converting text into tokens
print (tokens)      #printing
```

```python
freq = nltk.FreqDist(tokens)     #finding frequency of each token
for key,val in freq.items():     #printing token & it's frequency
    print (str(key) + ':' + str(val))
```

```python
freq.plot(20, cumulative=False)     #plotting graph between token and it's frequency
plt.show()       #displaying plot
```
